```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import scale
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
def load_data(filename):
    """
    Function loads data stored in the file filename and returns it as a numpy ndarray.
    
    Inputs:
        filename: given as a string.
        
    Outputs:
        Data contained in the file, returned as a numpy ndarray
    """
    return np.loadtxt(filename, skiprows=1,delimiter=' ')
```

```python
Xtrain = load_data('training_data.txt')[:,1:]
ytrain = load_data('training_data.txt')[:,0]
Xtest = load_data('test_data.txt')
```

```python
alpha = [0.01,0.1,1,10,100,500,1000,5000,10000]
avg = np.array([])
for i in alpha:
    clf = RidgeClassifier(alpha=i,solver='auto') 
    scores = cross_val_score(clf, Xtrain, ytrain, cv=5, scoring='accuracy')
    avg = np.append(avg,np.mean(scores))

plt.figure(1)
plt.plot(alpha,avg)

plt.title('cross validation on alpha 0.01 to 10000')
plt.xlabel('alpha')
plt.ylabel('cv score')
plt.show()

alpha = [90,100,110,120,130,140,150]
avg = np.array([])
for i in alpha:
    clf = RidgeClassifier(alpha=i,solver='auto') 
    scores = cross_val_score(clf, Xtrain, ytrain, cv=5,scoring='accuracy')
    avg = np.append(avg,np.mean(scores))

plt.figure(2)
plt.plot(alpha,avg)
plt.title('cross validation on alpha 90 to 150')
plt.xlabel('alpha')
plt.ylabel('cv score')
plt.show()
```
