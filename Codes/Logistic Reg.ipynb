```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
def load_data(filename):
    """
    Function loads data stored in the file filename and returns it as a numpy ndarray.
    
    Inputs:
        filename: given as a string.
        
    Outputs:
        Data contained in the file, returned as a numpy ndarray
    """
    return np.loadtxt(filename, skiprows=1,delimiter=' ')
```

```python
Xtrain = load_data('training_data.txt')[:,1:]
ytrain = load_data('training_data.txt')[:,0]
Xtest = load_data('test_data.txt')
```

```python
C = [0.01,0.1,1,10,50,100]
avg = np.array([])
for i in C:
    clf = LogisticRegression(C=i) 
    scores = cross_val_score(clf, Xtrain, ytrain, cv=5)
    avg = np.append(avg,np.mean(scores))

plt.figure(1)
plt.plot(C,avg)

plt.title('cross validation on C 0.01 to 100')
plt.xlabel('C')
plt.ylabel('cv score')
plt.show()

C = [0.05,0.06,0.07,0.08,0.09,1,1.01,1.02,1.03]
avg = np.array([])
for i in C:
    clf = LogisticRegression(C=i) 
    scores = cross_val_score(clf, Xtrain, ytrain, cv=5)
    avg = np.append(avg,np.mean(scores))

plt.figure(2)
plt.plot(C,avg)
plt.title('cross validation on C 0.05 to 1.03')
plt.xlabel('C')
plt.ylabel('cv score')
plt.show()
```
